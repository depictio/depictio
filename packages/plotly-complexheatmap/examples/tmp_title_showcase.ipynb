# Title & Description Showcase

Three examples demonstrating the new `title` and `description` parameters across different heatmap configurations.

```python
import numpy as np
import pandas as pd
from plotly_complexheatmap import ComplexHeatmap, HeatmapAnnotation

rng = np.random.default_rng(42)
```

## 1. Small heatmap — title only, no annotations

```python
# 8x6 matrix, no clustering, no annotations
small_data = rng.standard_normal((8, 6))
small_data[:4, :3] += 2.0

df_small = pd.DataFrame(
    small_data,
    index=[f"gene_{i}" for i in range(8)],
    columns=[f"S{j}" for j in range(6)],
)

hm1 = ComplexHeatmap(
    df_small,
    cluster_rows=False,
    cluster_cols=False,
    title="Small Heatmap",
    colorscale="Blues",
    width=500,
    height=400,
)
fig1 = hm1.to_plotly()
fig1.show()
```

## 2. Medium heatmap — title + description, with annotations

```python
# 30x12 matrix with row/column annotations
n_genes, n_samples = 30, 12
med_data = rng.standard_normal((n_genes, n_samples))
med_data[:10, :6] += 2.5
med_data[10:20, 6:] += 1.8

df_med = pd.DataFrame(
    med_data,
    index=[f"gene_{i:02d}" for i in range(n_genes)],
    columns=[f"sample_{j:02d}" for j in range(n_samples)],
)

# Column annotation: treatment group
groups = ["Control"] * 6 + ["Treated"] * 6
top_ha = HeatmapAnnotation(group=groups)

# Row annotation: pathway
pathways = ["Metabolism"] * 10 + ["Signaling"] * 10 + ["Other"] * 10
right_ha = HeatmapAnnotation(pathway=pathways, which="row")

hm2 = ComplexHeatmap(
    df_med,
    cluster_rows=True,
    cluster_cols=True,
    top_annotation=top_ha,
    right_annotation=right_ha,
    colorscale="RdBu_r",
    normalize="row",
    name="z-score",
    title="Gene Expression Analysis",
    description="Clustered heatmap with treatment groups and pathway annotations",
    width=900,
    height=700,
)
fig2 = hm2.to_plotly()
fig2.show()
```

## 3. Large heatmap — title + description, multiple annotation tracks

```python
# 60x20 matrix with multiple annotation tracks
n_genes, n_samples = 60, 20
big_data = rng.standard_normal((n_genes, n_samples))
big_data[:20, :7] += 3.0
big_data[20:40, 7:14] += 2.0
big_data[40:, 14:] += 2.5

df_big = pd.DataFrame(
    big_data,
    index=[f"gene_{i:03d}" for i in range(n_genes)],
    columns=[f"sample_{j:02d}" for j in range(n_samples)],
)

# Column annotations: batch + quality score
batch = ["Batch_A"] * 7 + ["Batch_B"] * 7 + ["Batch_C"] * 6
quality = rng.uniform(0.6, 1.0, size=n_samples)
top_ha = HeatmapAnnotation(batch=batch, quality=quality)

# Row annotations: category + numeric score
category = ["Apoptosis"] * 20 + ["Repair"] * 20 + ["Growth"] * 20
score = rng.uniform(0, 10, size=n_genes)
right_ha = HeatmapAnnotation(category=category, score=score, which="row")

hm3 = ComplexHeatmap(
    df_big,
    cluster_rows=True,
    cluster_cols=True,
    top_annotation=top_ha,
    right_annotation=right_ha,
    colorscale="Viridis",
    normalize="row",
    name="expression",
    title="Multi-Batch Expression Profiling",
    description="60 genes across 20 samples with batch and pathway annotations",
    width=1100,
    height=900,
)
fig3 = hm3.to_plotly()
fig3.show()
```
